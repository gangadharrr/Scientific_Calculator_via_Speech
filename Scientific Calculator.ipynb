```python
import nltk
from nltk.corpus import stopwords
import re
import speech_recognition as sr
import pyttsx3
import math
stop=stopwords.words('english')
```

```python
def convert_to_numbers(s):
    words_to_numbers = {
        'one': '1',
        'two': '2',
        'to':'2',
        'three': '3',
        'four': '4',
        'five': '5',
        'six': '6',
        'seven': '7',
        'eight': '8',
        'nine': '9',
        'zero': '0',
        'plus':'+',
        'minus':'-',
        'dash':'-',
        'add':'+',
        'multiplied by':'*',
        'cross':'*',
        'power':'**',
        '^':'**',
        'square':'**2',
        'cube':'**3',
        'divided by':'/',
        'equal to':'=',
        'equals to':'=',
        'equals':'=',
        'x':'*',
        'X':'*',
        'into':'*',
        'degrees':'',
        'sign':'sin',
        'route':'root',
        'routine' :'root 10',
        'cause':'cos',
        '°':''
    }
   
    
    pattern = re.compile(r'\b(' + '|'.join(words_to_numbers.keys()) + r')\b')
    s=re.sub(pattern, lambda x: words_to_numbers.get(x.group(),''), s)
    pattern = re.compile(r'\b(' + '|'.join(stop) + r')\b')
    s=re.sub(pattern, '', s)
    s=s.replace('  ',' ' )
    pattern = re.compile(r"√[0-9]+(?:\.[0-9]+)*")
    s=re.sub(pattern,lambda x: str(eval(x.group().replace('√',''))**(0.5)), s)
    pattern = re.compile(r'\s?sin\s?[0-9]+(?:\.[0-9]+)*\s?')
    s=re.sub(pattern, lambda x: "sin({0})".format(*re.findall('[0-9]+(?:\.[0-9]+)*',x.group())), s)
    pattern = re.compile(r'\s?cos\s?[0-9]+(?:\.[0-9]+)*\s?')
    s=re.sub(pattern, lambda x: "cos({0})".format(*re.findall('[0-9]+(?:\.[0-9]+)*',x.group())), s)
    pattern = re.compile(r'\s?tan\s?[0-9]+(?:\.[0-9]+)*\s?')
    s=re.sub(pattern, lambda x: "tan({0})".format(*re.findall('[0-9]+(?:\.[0-9]+)*',x.group())), s)
    pattern = re.compile(r'\s?log\s[0-9]+\sbase\s[0-9]+\s?')
    s=re.sub(pattern, lambda x: "logX({0},{1})".format(re.findall('[0-9]+(?:\.[0-9]+)*',x.group())[0],re.findall('[0-9]+(?:\.[0-9]+)*',x.group())[1]), s)
    pattern = re.compile(r'\s?log\s?[0-9]+(?:\.[0-9]+)*\s?')
    s=re.sub(pattern, lambda x: "log({0})".format(*re.findall('[0-9]+(?:\.[0-9]+)*',x.group())), s)
    pattern = re.compile(r'\s?root\s[0-9]+\sbase\s[0-9]+\s?')
    s=re.sub(pattern, lambda x: "root({0},{1})".format(re.findall('[0-9]+(?:\.[0-9]+)*',x.group())[0],re.findall('[0-9]+(?:\.[0-9]+)*',x.group())[1]), s)
    pattern = re.compile(r'\s?root\s\d+\s?')
    s = re.sub(pattern, lambda x: "root({0},2)".format(re.findall(r'\d+(?:\.\d+)*', x.group())[0]), s)
    pattern = re.compile(r"\b(?!sin|cos|tan|log|root|[0-9]+(?:\.[0-9]+)*|\s\b)\w+")
    s=re.sub(pattern,lambda x:'', s)
    
    return s

def primitive_replacement(s):
    words_to_numbers = {
        'audition': 'addition',
        'additional': 'addition',
        
    }
    pattern = re.compile(r'\b(' + '|'.join(words_to_numbers.keys()) + r')\b')
    s=re.sub(pattern, lambda x: words_to_numbers.get(x.group(),''), s)
    return s
```

```python
add=lambda x,y:x+y
sub=lambda x,y:x-y
mul=lambda x,y:x*y
div=lambda x,y:x/y
sin=lambda x:math.sin(math.radians(x))
cos=lambda x:math.cos(math.radians(x))
tan=lambda x:math.tan(math.radians(x))
log=lambda x:math.log(x)
logX=lambda *x :math.log(x[0],x[1])
root=lambda *x:(x[0])**(1/x[1])
root2=lambda x:(x)**(1/2)
root3=lambda x:(x)**(1/3)
    
```

```python
r = sr.Recognizer()
def SpeakText(command):
    engine = pyttsx3.init()
    voice = engine.getProperty('voices')
    engine.setProperty('voice', voice[1].id) 
    engine.say(command)
    engine.runAndWait()
     
while(1):  
    try:
        with sr.Microphone() as source2:
            r.adjust_for_ambient_noise(source2, duration=0.5)
            audio2 = r.listen(source2,phrase_time_limit=7)
            MyText = r.recognize_google(audio2,show_all=False)
            #MyText = MyText
            print(MyText)
            SpeakText(MyText)
    except sr.RequestError as e:
        print("Could not request results; {0}".format(e))
    except sr.UnknownValueError:
        print(1)
```

```python
r = sr.Recognizer()
def SpeakText(command):
    engine = pyttsx3.init()
    engine.say(command)
    engine.runAndWait()
     
while(1):  
    try:
        with sr.Microphone() as source2:
            r.adjust_for_ambient_noise(source2, duration=0.5)
            audio2 = r.listen(source2,phrase_time_limit=7)
            MyText = r.recognize_google(audio2)
            MyText = MyText.lower()
            print(MyText)
            MyText=primitive_replacement(MyText).replace('°','')
            _MyText=MyText
            if(MyText.find('addition')!=-1):
                try:
                    MyText=MyText[MyText.find('addition'):]
                    MyText=list(convert_to_numbers(MyText).split())
                    print('addition',MyText)
                    print(f"Result of '{_MyText}' is : {eval(MyText[-2])+eval(MyText[-1])}")
                except Exception as e:
                    print(e)
                    print("!!!!Try Again!!!!")
                
            elif(MyText.find('subtraction')!=-1):
                try:
                    MyText=MyText[MyText.find('subtraction'):]
                    MyText=list(convert_to_numbers(MyText).split())
                    print('subtraction',MyText)
                    print(f"Result of '{_MyText}' is : {eval(MyText[-2])-eval(MyText[-1])}")
                except Exception as e:
                    print(e)
                    print("!!!!Try Again!!!!")
                
            elif(MyText.find('multiplication')!=-1):
                try:
                    MyText=MyText[MyText.find('multiplication'):]
                    MyText=list(convert_to_numbers(MyText).split())
                    print('multiplication',MyText)
                    print(f"Result of '{_MyText}' is : {eval(MyText[-2])*eval(MyText[-1])}")
                except Exception as e:
                    print(e)
                    print("!!!!Try Again!!!!")
                
            elif(MyText.find('division')!=-1):
                try:
                    MyText=MyText[MyText.find('division'):]
                    MyText=list(convert_to_numbers(MyText).split())
                    print('division',MyText)
                    print(f"Result of '{_MyText}' is : {eval(MyText[-2])/eval(MyText[-1])}")
                except Exception as e:
                    print(e)
                    print("!!!!Try Again!!!!")
                
            elif(MyText.find('cube root')!=-1):
                try:
                    MyText=MyText[MyText.find('cube root'):]
                    MyText=list(convert_to_numbers(MyText).split())
                    pattern = re.compile(r'[^\d]')
                    MyText[-1]=re.sub(pattern, lambda x: '', MyText[-1])
                    print('cube root',MyText)
                    print(f"Result of '{_MyText}' is : {root(eval(MyText[-1]),3)}")
                except Exception as e:
                    print(e)
                    print("!!!!Try Again!!!!")
                
            elif(MyText.find('square root')!=-1):
                try:
                    MyText=MyText[MyText.find('square root'):]
                    MyText=list(convert_to_numbers(MyText).split())
                    pattern = re.compile(r'[^\d]')
                    MyText[-1]=re.sub(pattern, lambda x: '', MyText[-1])
                    print('square root',MyText)
                    print(f"Result of '{_MyText}' is : {root(eval(MyText[-1]),3)}")
                except Exception as e:
                    print(e)
                    print("!!!!Try Again!!!!")
            
            elif(MyText.find('evaluate')!=-1):
                try:
                    MyText=MyText[MyText.find('evaluate'):]
                    MyText=list(convert_to_numbers(MyText).split("equation"))
                    print('evaluate',MyText)
                    print(f"Result of '{_MyText}' is : {eval(MyText[-1])}")
                except Exception as e:
                    print(e)
                    print("!!!!Try Again!!!!")
            else:
                try:
                    MyText=[i for i in convert_to_numbers(MyText).split() if not i.isalpha()]
                    print(MyText)
                    print(f"Result of '{_MyText}' is : {eval(''.join(MyText))}")
                except Exception as e:
                    print(e)
                    print("!!!!Try Again!!!!")
  
    except sr.RequestError as e:
        print("Could not request results; {0}".format(e))
    except sr.UnknownValueError:
        print(1)
```
